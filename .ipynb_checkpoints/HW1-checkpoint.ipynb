```python
%matplotlib inline 
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from numpy import ma

font = {'family' : 'normal',
        'weight' : 'normal',
        'size'   : 22}

matplotlib.rc('font', **font)

k = 9E9
q = 1.6E-19


X, Y = np.meshgrid(np.arange(-10, 10, .5), np.arange(-10, 10, .5))

#Components of the E Field for a single charge
EX = k*q*X/(np.sqrt((X)**2+(Y)**2))**3
EY = k*q*(Y)/(np.sqrt((X)**2+(Y)**2))**3


# 1
plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')
Q = plt.quiver(X, Y, EX, EY)

plt.title('Positive Point Charge')



plt.show()
```

```python
#Components of the E Field for a dipole, the positive charge at y = 3 and the
#negative charge at y = -3
EX_up = k*q*X/(np.sqrt((X)**2+(Y-3)**2))**3
EY_up = k*q*(Y-3)/(np.sqrt((X)**2+(Y-3)**2))**3
EX_down = -k*q*X/(np.sqrt((X)**2+(Y+3)**2))**3
EY_down = -k*q*(Y+3)/(np.sqrt((X)**2+(Y+3)**2))**3

# 2
plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')
Q = plt.quiver(X, Y, EX_up + EX_down, EY_up + EY_down)

plt.title('Dipole Centered at the Origin')
plt.show()
```

```python
#Components for a quadrupole
EX_u_r = -k*q*(X-3)/(np.sqrt((X-3)**2+(Y-3)**2))**3
EY_u_r = -k*q*(Y-3)/(np.sqrt((X-3)**2+(Y-3)**2))**3

EX_d_l = -k*q*(X+3)/(np.sqrt((X+3)**2+(Y+3)**2))**3
EY_d_l = -k*q*(Y+3)/(np.sqrt((X+3)**2+(Y+3)**2))**3

EX_u_l = k*q*(X+3)/(np.sqrt((X+3)**2+(Y-3)**2))**3
EY_u_l = k*q*(Y-3)/(np.sqrt((X+3)**2+(Y-3)**2))**3

EX_d_r = k*q*(X-3)/(np.sqrt((X-3)**2+(Y+3)**2))**3
EY_d_r = k*q*(Y+3)/(np.sqrt((X-3)**2+(Y+3)**2))**3

EX_quad = EX_u_r + EX_d_l + EX_u_l + EX_d_r
EY_quad = EY_u_r + EY_d_l + EY_u_l + EY_d_r

# 3
plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')
Q = plt.quiver(X, Y, EX_quad, EY_quad)

plt.title('Quadrupole Centered at the Origin')
plt.show()
```
